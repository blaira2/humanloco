```python
#build diffusion
import numpy as np
import torch
from torch.utils.data import DataLoader
from diffusion import HumanoidDiffusionDataset, PoseDiffusion, morph_vector_from_params
from humanoid_module import VARIANTS

def build_diffusion_dataset(traj_files, past_len=10, future_len=10):
    episodes = []
    for name, cfg in VARIANTS.items():
        file = f"{name}_traj.npz"
        if file not in traj_files:
            continue
        data = np.load(file)
        qpos = data["qpos"]  # [T, D_q]
        morph_vec = morph_vector_from_params(cfg)  # uses MORPH_KEYS in diffusion.py
        episodes.append({"qpos": qpos, "morph": morph_vec})

    dataset = HumanoidDiffusionDataset(
        episodes=episodes,
        past_len=past_len,
        future_len=future_len
    )
    return dataset

traj_files = [f"{name}_traj.npz" for name in VARIANTS.keys()]
dataset = build_diffusion_dataset(traj_files, past_len=10, future_len=10)
loader = DataLoader(dataset, batch_size=64, shuffle=True)
```

```python
#Train diffusion

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

diffusion_model = PoseDiffusion(
    D_target=dataset.D_target,
    D_cond=dataset.D_cond,
    n_steps=100,
    hidden=512
).to(device)

opt = torch.optim.Adam(diffusion_model.parameters(), lr=1e-4)

def train_diffusion(num_epochs=20):
    diffusion_model.train()
    for epoch in range(num_epochs):
        total_loss = 0.0
        for x_0, cond in loader:
            x_0 = x_0.to(device)      # future pose chunk flattened
            cond = cond.to(device)    # past poses + morphology

            B = x_0.size(0)
            t = torch.randint(
                0, diffusion_model.n_steps,
                (B,), device=device, dtype=torch.long
            )
            noise = torch.randn_like(x_0)
            x_t = diffusion_model.q_sample(x_0, t, noise=noise)

            eps_pred = diffusion_model(x_t, t, cond)
            loss = torch.mean((eps_pred - noise) ** 2)

            opt.zero_grad()
            loss.backward()
            opt.step()

            total_loss += loss.item() * B

        print(f"Epoch {epoch+1}: loss = {total_loss / len(dataset):.6f}")

    torch.save(diffusion_model.state_dict(), "pose_diffusion.pt")

train_diffusion()
```
